```python
from pickle import load
import matplotlib.pyplot as plt
from keras.models import load_model
from keras.preprocessing.image import ImageDataGenerator
import keras
import tensorflow as tf
from keras.layers.core import Lambda
```

```python
from keras.datasets import cifar10
from keras.utils import to_categorical
# load data
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
```

## Result of ARL92

```python
model_name = "ARL92"
with open('Result/{}/History/train_lossHistory'.format(model_name), 'rb') as handle: # loading old history 
    train_lossHistory = load(handle)
with open('Result/{}/History/train_accHistory'.format(model_name), 'rb') as handle: # loading old history 
    train_accHistory = load(handle)
with open('Result/{}/History/vali_lossHistory'.format(model_name), 'rb') as handle: # loading old history 
    vali_lossHistory = load(handle)
with open('Result/{}/History/vali_accHistory'.format(model_name), 'rb') as handle: # loading old history 
    vali_accHistory = load(handle)

    
# Plot training & validation accuracy values
plt.plot(train_lossHistory)
plt.plot(vali_lossHistory)
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.show()

# Plot training & validation loss values
plt.plot(train_accHistory)
plt.plot(vali_accHistory)
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.show()
```
